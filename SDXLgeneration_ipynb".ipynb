```python
#импорт библиотек
from diffusers import StableDiffusionXLPipeline
from diffusers.utils import make_image_grid
import torch
```

# Код

```python
#установка модели
pipe = StableDiffusionXLPipeline.from_pretrained("segmind/SSD-1B",    torch_dtype=torch.float16, use_safetensors=True, variant="fp16")
pipe.to("cuda")
```

```python
prompt = "drawing of 2 cats white with ginger and black sitting on a sofa in a large living room watching TV" # Позитивный промпт
neg_prompt = "" # Негативный промпт

images = pipe(prompt=prompt, negative_prompt=neg_prompt, width=512, height=256, num_images_per_prompt=4).images
make_image_grid(images, rows=2, cols=2)
```
